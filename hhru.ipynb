```python
import requests
import pandas as pd
from time import sleep
import plotly.express as px
```

```python
url = 'https://api.hh.ru/vacancies'
```

```python
params = {
    'text': '',
    'area': 66,
    'per_page': 100,
    'page': 0
}
```

```python
all_vacancies = []
```

```python
max_pages = 40
```

```python
for page in range(max_pages):
  params['page'] = page
  response = requests.get(url, params=params)

  if response.status_code == 200:
    data = response.json()
    vacancies = data.get('items', [])
    all_vacancies.extend(vacancies)

  sleep(0.5)
```

```python
df = pd.DataFrame(all_vacancies)
```

```python
def extract_salary_info(salary_dict):
    if not isinstance(salary_dict, dict):  # Проверка, что это словарь
        return {'salary_from': None, 'salary_currency': None, 'salary_type': None}

    # Извлекаем нужные поля
    salary_from = salary_dict.get('from')
    salary_currency = salary_dict.get('currency')
    salary_type = 'Gross' if salary_dict.get('gross') else 'Net'

    return {
        'salary_from': salary_from,
        'salary_currency': salary_currency,
        'salary_type': salary_type
    }
```

```python
salary_info_df = df['salary'].apply(extract_salary_info).apply(pd.Series)
```

```python
df = pd.concat([df, salary_info_df], axis=1)
```

```python
df = df.drop('salary', axis=1)
```

```python
def xtract(w_dict):
  if not isinstance(w_dict, dict):
        return {'id_employer': None, 'name_employer': None, 'accredited_it_employer': None}

  id_employer = w_dict.get('id')
  name_employer = w_dict.get('name')
  accredited_it_employer = w_dict.get('accredited_it_employer')

  return {
        'id_employer': id_employer,
        'name_employer': name_employer,
        'accredited_it_employer': accredited_it_employer
    }
```

```python
def extract_metro_station(location_dict):
    if not isinstance(location_dict, dict):
        return "Данные отсутствуют"

    if location_dict.get('metro') and isinstance(location_dict['metro'], dict):
        return location_dict['metro'].get('station_name', "Данные отсутствуют")

    return "Данные отсутствуют"
```
